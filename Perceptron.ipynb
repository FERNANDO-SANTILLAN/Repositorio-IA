```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Datos de entrenamiento
# Clase 1 (etiqueta 0)
x1 = np.random.randn(50, 2) - 2
y1 = np.zeros(50)
```

```python
# Clase 2 (etiqueta 1)
x2 = np.random.randn(50, 2) + 2
y2 = np.ones(50)
```

```python
# Combinar datos
X = np.vstack((x1, x2))
y = np.hstack((y1, y2))
```

```python
# Sesgo (bias) como columna de 1's
X_bias = np.hstack((np.ones((X.shape[0], 1)), X))
```

```python
# Inicializar pesos
pesos = np.random.randn(3)
```

```python
# Activación escalón
def activacion(x):
    return 1 if x >= 0 else 0
```

```python
# Entrenamiento del perceptrón
tasa_aprendizaje = 0.1
epocas = 10

for epoca in range(epocas):
    for i in range(len(X_bias)):
        xi = X_bias[i]
        yi = y[i]
        salida = activacion(np.dot(xi, pesos))
        error = yi - salida
        pesos += tasa_aprendizaje * error * xi

print("Pesos entrenados:", pesos)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x1[:, 0], x1[:, 1], color='red', label='Clase 0')
plt.scatter(x2[:, 0], x2[:, 1], color='blue', label='Clase 1')

# línea de decisión
x_vals = np.linspace(-6, 6, 100)
y_vals = -(pesos[0] + pesos[1] * x_vals) / pesos[2]
plt.plot(x_vals, y_vals, color='green', label='Línea de decisión')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptrón")
plt.legend()
plt.grid(True)
plt.show()
```
