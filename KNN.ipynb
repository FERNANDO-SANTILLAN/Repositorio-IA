```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data
y = iris.target
print("Nombres de las clases:", iris.target_names)
print("Caracter√≠sticas:", iris.feature_names)
```

```python
# Dos caracter√≠sticas para graficar (por ejemplo, largo y ancho del p√©talo)
plt.figure(figsize=(8,6))
for i in range(3):
    plt.scatter(X[y == i, 2], X[y == i, 3], label=iris.target_names[i])
plt.xlabel("Largo del p√©talo (cm)")
plt.ylabel("Ancho del p√©talo (cm)")
plt.title("Visualizaci√≥n de clases Iris üå∏")
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Modelo KNN con k=3
modelo_knn = KNeighborsClassifier(n_neighbors=3)
modelo_knn.fit(X_entrenamiento, y_entrenamiento)
```

```python
# Predecir sobre los datos de prueba
y_pred = modelo_knn.predict(X_prueba)
print("Exactitud:", accuracy_score(y_prueba, y_pred))
print("\nReporte de clasificaci√≥n:\n", classification_report(y_prueba, y_pred, target_names=iris.target_names))
```

```python
# Visualizaci√≥n de los datos de prueba con sus predicciones
plt.figure(figsize=(8,6))
for i in range(3):
    plt.scatter(X_prueba[y_pred == i, 2], X_prueba[y_pred == i, 3], label=f"Predicho: {iris.target_names[i]}")
plt.xlabel("Largo del p√©talo (cm)")
plt.ylabel("Ancho del p√©talo (cm)")
plt.title("Predicciones del modelo KNN")
plt.legend()
plt.grid(True)
plt.show()
```
