```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_informative=2,
                           n_redundant=0, n_clusters_per_class=1, n_classes=2, random_state=42)
```

```python
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Clasificador KNN con k=5 vecinos
modelo_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# Entrenamiento del modelo con los datos de entrenamiento
modelo_knn.fit(X_entrenamiento, y_entrenamiento)
```

```python
# Predicciones
y_prediccion = modelo_knn.predict(X_prueba)
```

```python
# Precisión del modelo
precision = accuracy_score(y_prueba, y_prediccion)
print(f"Precisión del modelo: {precision:.2f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_prueba[:, 0], X_prueba[:, 1], c=y_prediccion, cmap='coolwarm', edgecolor='k', s=100)
plt.title('Clasificación KNN (K=5)')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.grid(True)
plt.show()
```
